# Automação de Sistemas e Processos com Python

### Desafio:

Para controle de custos, todos os dias, seu chefe pede um relatório com todas as compras de mercadorias da empresa.
O seu trabalho, como analista, é enviar um e-mail para ele, assim que começar a trabalhar, com o total gasto, a quantidade de produtos compradas e o preço médio dos produtos.

E-mail do seu chefe: para o nosso exercício, coloque um e-mail seu como sendo o e-mail do seu chefe<br>
Link de acesso ao sistema da empresa: https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
import pyautogui
import time
import pyperclip
# passo a passo
pyautogui.pause = 0.5
# passo 1: Entrar no sistema da empresa (no link)
pyautogui.hotkey ("win")
pyautogui.write ("https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema")
pyautogui.press ("enter")

# tempo de carregar os comandos
time.sleep(5)

#   pyautogui.click #clilcar com o mouse
#   pyautogui.write #escrever um texto
#   pyautogui.press #apertar uma tecla
#   pyautogui.hotkey #apertar uma combiação de teclas


# passo 2: fazer login
pyautogui.click (x=713, y=375)
pyautogui.write ("meu_login")
time.sleep(1)
# Senha
pyautogui.click (x=654, y=455)
pyautogui.write ("minha_senha")

#confirmação de enter
pyautogui.click (x=671, y=530)

# passo 3: exportar a base de dadoos (download o arquivo)
#pyautogui.click (x=518, y=330)
#pyautogui.click (x=1661, y=330)
#pyautogui.click (x=1661, y=330)
#time.sleep(3) #espera do download

# passo 4: calcular os indicadores
import pandas as pd

tabela = pd.read_csv(r"https://drive.google.com/file/d/1yG2dHcFGLg9kXE5i5nfWSvdOHSOz4k41/view?usp=share_link", sep=";")
display(tabela)

total_gasto = tabela["ValorFinal"].sum()
quantidade = tabela["Quantidade"].sum()
preco_medio = total_gasto / quantidade
print (total_gasto)
print (quantidade)
print (preco_medio)

# passo 5: enivar o email para meu chefe
#   pyautogui.click #clilcar com o mouse
#   pyautogui.write #escrever um texto
#   pyautogui.press #apertar uma tecla
#   pyautogui.hotkey #apertar uma combiação de teclas


pyautogui.hotkey ("ctrl", "t")
pyautogui.write("https://mail.google.com/mail/u/0/")
pyautogui.press("enter")

time.sleep(5)

#cliclar no botão escrever
pyautogui.click (x=71, y=195)
#peencher as informações do email
time.sleep(5)
#destinatario
pyautogui.write('@gmail.com') #email destinatario
pyautogui.press("tab")
# escolher destinario
pyautogui.press("tab")
time.sleep(5) 
#passar para o campo assunto
pyautogui.write('Relatorio de Vendas')


pyautogui.press("tab") 
#passando para o corpo do email

texto = f"""
prezados,
segue o relatorio de compras

total de gasto: R${total_gasto:,.2f}
quantidade de produtos: R${quantidade:,}
preço medio: R${preco_medio:,.2f}

Qualquer duvida, é só falar
att, Angelo python
"""

pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v") 

#enviar

pyautogui.hotkey("ctrl", "enter")

```
